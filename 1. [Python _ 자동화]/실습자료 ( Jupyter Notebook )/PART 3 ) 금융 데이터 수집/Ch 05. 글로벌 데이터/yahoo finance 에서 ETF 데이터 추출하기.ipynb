```python
payload = {'symbol' : 'IWM',
'period1' :'1483196400',
'period2' : '1563548400',
'interval' : '1d',
'includePrePost' : 'true',
'events' : 'div|split|earn',
'lang' : 'en-US',
'region' : 'US',
'corsDomain' :'finance.yahoo.com'}
```

```python
ticker = 'GSC'
```

```python
payload['symbol'] = ticker
```

```python
url = "https://query1.finance.yahoo.com/v8/finance/chart/{}?"
```

```python
r = requests.get(url.format(ticker), params=payload)
```

```python
dict_rt = json.loads(r.text)
```

```python
dict_rt
```

```python
dict_rt['chart']['result'][0].keys()
```

```python
dict_rt['chart']['result'][0]['timestamp']
```

```python
dict_rt['chart']['result'][0]['indicators']['quote'][0]
```

```python
dict_rt['chart']['result'][0]['indicators']['quote'][0].keys()
```

```python
data = {"date" : dict_rt['chart']['result'][0]['timestamp'],
 "close" : dict_rt['chart']['result'][0]['indicators']['quote'][0]['close'],
 "open"  : dict_rt['chart']['result'][0]['indicators']['quote'][0]['open'],
 "high"  : dict_rt['chart']['result'][0]['indicators']['quote'][0]['high'],
 "low"   : dict_rt['chart']['result'][0]['indicators']['quote'][0]['low'],
 'volume' : dict_rt['chart']['result'][0]['indicators']['quote'][0]['volume']}
```

```python
df = pd.DataFrame(data)
```
