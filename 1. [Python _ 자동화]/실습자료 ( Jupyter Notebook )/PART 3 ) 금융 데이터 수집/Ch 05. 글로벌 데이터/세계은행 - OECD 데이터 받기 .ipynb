# World Bank 

* GDP(현재 US$)		NY.GDP.MKTP.CD	GDP (current US$)
* 1인당 	GNI, 아틀라스 방식(현재 US$)	NY.GNP.ATLS.CD	GNI, Atlas method (current US$)
* 총인구		SP.POP.TOTL	Population, total
* CO2 배출(1인당 미터톤)	EN.ATM.CO2E.PC	CO2 emissions (metric tons per capita)
* 출생시 총 기대수명(년) SP.DYN.LE00.IN	Life expectancy at birth, total (years)

```python
Indicator_code = ['NY.GDP.MKTP.CD', 'NY.GNP.ATLS.CD', 'SP.POP.TOTL', 'EN.ATM.CO2E.PC', 'SP.DYN.LE00.IN' ]
```

```python
world_url = "http://api.worldbank.org/v2/country/{}/indicator/{}/?downloadformat=excel&dataformat=table"
```

```python
total = []
for country_code in oecd_code:
    country_df = pd.read_excel(world_url.format(country_code.lower(), 'NY.GDP.MKTP.CD'))
    country_df.drop(country_df.index[:2], inplace=True)
    country_df_trans = country_df.transpose()
    column_name = country_df_trans.iloc[2,1]
    country_df_trans.columns = ['year', column_name]
    country_df_trans.drop(country_df_trans.index[:4], inplace=True)
    country_df_trans.reset_index(drop=True, inplace=True)
    country_df_trans['country'] = country_code
    total.append(country_df_trans)
```

```python
pd.concat(total)
```

```python
total = []
for code in Indicator_code:
    for country_code in oecd_code:
        print(country_code)
        country_df = pd.read_excel(world_url.format(country_code.lower(), code))
        country_df.drop(country_df.index[:2], inplace=True)
        country_df_trans = country_df.transpose()
        column_name = country_df_trans.iloc[2,1]
        country_df_trans.columns = ['year', 'value']
        country_df_trans.drop(country_df_trans.index[:4], inplace=True)
        country_df_trans.reset_index(drop=True, inplace=True)
        country_df_trans['country'] = country_code
        country_df_trans['indicator_code'] = code
        total.append(country_df_trans)
```

```python
oecd_df = pd.concat(total)
```
