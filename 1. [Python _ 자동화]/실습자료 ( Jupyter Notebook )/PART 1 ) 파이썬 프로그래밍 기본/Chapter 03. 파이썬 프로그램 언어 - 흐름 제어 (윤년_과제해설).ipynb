# 윤년 계산하기 

## 윤년의 조건 
* 연수가 4로 나누어 떨어지는 해는 윤년으로 한다. (e.g. 2004, 2008, 2012, 2016, 2020)
* 이중에서 100으로 나누어 떨어지는 해는 평년으로 한다.(1900, 2100, 2200, 2300)
* 그 중에서 400으로 나누어 떨어지는 해는 윤년으로 둔다. 

* 위의 3가지 규칙을 사용하여 1900년 ~ 2100년까지 윤년을 출력하시오

```python
# 아래 2가지 방법 모두 초기 리스트 생성
#leap_year = list()
leap_year = []

for year in range(1900, 2101):
    if(year % 4 == 0 and year % 100 != 0 or year % 400 == 0):
        leap_year.append(year)
```
