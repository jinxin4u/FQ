* popular series의 series 값을 수집해보겠습니다. 

```python
r= requests.get("https://fred.stlouisfed.org/tags/series?et=&ob=pv&pageID=1&t=")
```

```python
bs = BeautifulSoup(r.text, 'lxml')
```

```python
bs.find("a", {'title' : 'last page'}).text
```

```python
import re
```

```python
last = int(re.search('[0-9]+', bs.find("a", {'title' : 'last page'}).text).group()) + 1
```

```python
last
```

```python
series = []
for idx in range(1, last):
    r= requests.get("https://fred.stlouisfed.org/tags/series?et=&ob=pv&pageID={}&t=".format(idx))
    bs = BeautifulSoup(r.text, 'lxml')
    for content in bs.findAll("a", {'class':"series-title pager-series-title-gtm"}):
        print (content.text.strip(), "\t", content['href'])
        series.append([content.text.strip(),  content['href'].split('/')[-1]])
```

```python
series
```

```python
pd.DataFrame(series , columns=['title', 'series'])
```
