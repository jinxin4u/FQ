```python
total = []
state = 0

for idx in range(1, 220):
    print (idx)
    payload['pn'] = idx
    while(True):
        try:
            r = requests.post(url, data=payload, headers = header)
            rt =json.loads(r.text)
            rt['hits']
            cnt = 0
            break
        except KeyError:
            print ("KeyError가 발생했습니다.")
            print (r.status_code)
            if(cnt == 5):
                state = -1
                print ("KeyError 종료")
                break
            time.sleep(3)
            cnt += 1
        except Exception as erorr:
            print (error)
            if(cnt == 5):
                state = -1
                print (error, "종료")
                break
            time.sleep(3)
            cnt += 1
            
    
    if (state == -1):
        print ("프로그램 종료")
        break
            
    for x in rt['hits']:
        #print ( x['name_trans'], x['stock_symbol'], x['exchange_trans'], x['industry_trans'])
        total.append([x['name_trans'], x['stock_symbol'], x['exchange_trans'], x['industry_trans']])
```

```python
pd.DataFrame(total, columns=['name_trans', 'stock_symbol', 'exchange_trans', 'industry_trans'])
```
