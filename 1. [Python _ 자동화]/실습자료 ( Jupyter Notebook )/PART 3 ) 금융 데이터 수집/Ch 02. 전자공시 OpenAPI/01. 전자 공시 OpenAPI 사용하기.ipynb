# 전자 공시 OpenAPI 사용하기 
* 전자공시시스템(dart.fss.or.kr)에 공시되는 문서를 api로 확인해봅시다.
* 공시되는 문서를 다음 챕터에서 다운로드하고 읽는 방법을 공부하겠습니다.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

* dart api 주소를 api_url에 저장합니다.

```python
api_url = "http://dart.fss.or.kr/api/search.json?"
```

* 삼성전자 코드를 예시를 만들어보겠습니다.

```python
value = ['005930']
```

* 전달할 데이터를 dict 형태로 만듭니다. 해당 키 값들은 Dart의 api 사용법에 나오는 내용중에서 필요한 부분만 넣은 것입니다.
* auth는 강사의 api 키입니다. 회원 가입하고 본인 키를 사용해주세요 

```python
payload = {'auth' : '45eedf32874002b7d41884abba4ba52262d57c2d',
            'start_dt' : '20190101',
            'bsn_tp' : ['A002', 'A003', 'B001', 'D001'],
            'crp_cd' : value}
```

* requests의 get() 메소드를 사용할 때 param 매개변수에 위에서 만든 dict를 전달합니다. 
* 그러면 

```python
r = requests.get(api_url, params=payload)
```

* params으로 전달된 dict 값으로 아래 주소가 생성됩니다. 결국 아래 주소를 실행하는 것입니다.

```python
r.url
```

```python
r.status_code
```

* json형태로 보이는 str 객체를 dict 형태로 변경하기 위해서 json의 loads 메소드를 사용합니다.

```python
json.loads(r.text)
```

* 이전 수업 시간에 배웠던 거래소 사이트의 종목 코드를 읽어옵니다.
    > 한국거래소(KRX) - 전체종목 리스트 (종목검색기) 내용에서 저장을 csv로 한 결과입니다.
    
    > 지금 처음 보시는 면 첨부된 csv파일을 사용하세요

```python
with open("./code.csv", encoding='utf8') as f:
    company = f.read().split("\n")
```

* csv 파일을 개행문자("\n") 기준으로 split() 메소드로 분리합니다.

```python
company[2]
```

* company 리스트에는 각 회사에 대한 정보가 하나의 문자열로 존재하고 그 값을 ","로 split() 메소드로 분리합니다. 
* [-1] 마지막 리스트의 [1:] 두번째부터 끝까지 인덱싱을 합니다.

```python
company[1].split(",")[-1][1:]
```

```python
for idx, cnt in enumerate(company):
    print (idx, cnt.split(",")[-1][1:])
```

* 첫째행과 마지막행은 우리가 원하는 정보가 아니기에 [1:-1] 두번째부터 마지막전까지 인덱싱을 합니다.
* for 루프를 돌면서 회사 코드를 payload안에 입력하고 api를 실행하여 각 회사별 원하는 문서 정보를 리턴받습니다.
* 리턴 받은 정보를 json.loads를 사용하여 dict로 만들고 다시 DataFrame으로 최종 생성합니다. 
* 생성된 DataFrame를 total_df라는 DataFrame과 concat() 메소드를 사용하여 누적하여 합칩니다.

```python
total_df = pd.DataFrame()
for idx, cnt in enumerate(company[1:-1]):
    # 데이터가 너무 많으니 100개만 확인해보겠습니다.
    if(idx == 100):
        print (idx)
        break
    value = cnt.split(",")[-1][1:]
    payload = {'auth' : '45eedf32874002b7d41884abba4ba52262d57c2d',
            'start_dt' : '20190101',
            'bsn_tp' : ['I001'],
            'crp_cd' : value} # 회사명은 동적으로 루프를 돌면서 계속 바뀝니다. 
    r = requests.get(api_url, params=payload)
    # total_df에 전달받은 내용을 df로 만들고 합칩니다.
    total_df = pd.concat([total_df, pd.DataFrame(json.loads(r.text)['list']) ])
```
