```python
import requests
import json
import pandas as pd
```

```python
url = "http://ecos.bok.or.kr/api/KeyStatisticList/810YONO9D5OK5CJPXG3G/json/kr/1/1000/"
```

```python
r = requests.get(url)
```

* 한국은행 open api 중 통계 조회 조건 설정에 대해서 알아보겠습니다. 
* 나머지 서비스의 api는 동일한 방식으로 데이터를 가져올 수 있습니다.

```python
url = "http://ecos.bok.or.kr/api/StatisticSearch/{}/{}/{}/{}/{}/{}/{}/{}/{}/{}"
```

```python
key = "810YONO9D5OK5CJPXG3G"
type_ = 'json'
language = 'kr'
start = 1
end = 100
code = '064Y001'
period = 'DD' # MM
start_day = '20190101'
end_day = "20190719"
type_code = "0113000"
```

```python
url.format(key, type_, language, start,end, code, period, start_day, end_day, type_code)
```

```python
r = requests.get(url.format(key, type_, language, start,end, code, period, start_day, end_day, type_code))
```

```python
r.text
```

```python
pd.DataFrame(json.loads(r.text)['StatisticSearch']['row'])
```
