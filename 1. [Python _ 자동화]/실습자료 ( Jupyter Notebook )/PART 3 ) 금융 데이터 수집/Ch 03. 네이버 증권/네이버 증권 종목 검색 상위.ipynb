## 상승

```python
base_url = "https://finance.naver.com/sise/field_submit.nhn?"
```

* item 찾기

```python
url = "https://finance.naver.com/sise/sise_rise.nhn"
r = requests.get(url)
```

```python
bs = BeautifulSoup(r.text, "lxml")
```

```python
item_result = bs.find("table", class_='item_list').findAll("input")
```

```python
item_result[0]['value']
```

```python
item = [x['value'] for x in  item_result]
```

```python
item
```

```python
len(item)
```

* 데이터 가져 오기

```python
payload = {"menu" : 'rise',
           # sosok = 0 -> 코스피
           # sosok = 1 -> 코스닥
           'returnUrl' : 'http://finance.naver.com/sise/sise_rise.nhn?sosok=0',
           'fieldIds' : item[:7]}
```

```python
r = requests.get(base_url, params=payload)
bs = BeautifulSoup(r.text, "lxml")
```

```python
a = pd.read_html(str(bs.find("table", "type_2")))[0]
```

```python
item[:7]
```

```python
a.iloc[a['N'].dropna().index, :]
```

```python
rt_a = a.iloc[a['N'].dropna().index,:]
```

```python
rt_a.head(5)
```

```python
payload['fieldIds'] = item[8:15]
r = requests.get(base_url, params=payload)
bs = BeautifulSoup(r.text, "lxml")
```

```python
b = pd.read_html(str(bs.find("table", "type_2")))[0]
```

```python
rt_b = b.iloc[b['N'].dropna().index,:]
```
